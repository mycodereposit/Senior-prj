```python
start = max(load['datetime'].min(),pv['datetime'].min()) 
end = min(load['datetime'].max(),pv['datetime'].max())
t1 = 0
t2 = 0
t3 = 0
t4 = 0
load_mean = load[(start <= load['datetime']) & (load['datetime'] < end)]['PLtot'].mean()
pv_mean = pv[(start <= pv['datetime']) & (pv['datetime'] < end)]['PVtot'].mean()
dataset_detail = []
for i in range((end - start).days) :
    today = start + pd.Timedelta(i,unit='D')
    next_4_days = today + pd.Timedelta(4,unit='D')
    load_batch = load[(today <=load['datetime']) & (load['datetime'] < next_4_days)][['datetime','PLa','PLb','PLc','PLtot']]
    pv_batch = pv[(today <= pv['datetime']) & (pv['datetime'] < next_4_days)][['datetime','PVa','PVb','PVc','PVtot']]
    if load_batch.shape[0] == 384 and pv_batch.shape[0] == 384 :
        
        load_batch_mean  =  load_batch['PLtot'].mean()
        pv_batch_mean = pv_batch['PVtot'].mean()
        
        if load_batch_mean >= load_mean and pv_batch_mean >= pv_mean:
            load_type = 'high_load'
            pv_type = 'high_solar'
            t1 += 1
            k = t1
        elif load_batch_mean >= load_mean and pv_batch_mean < pv_mean:
            load_type = 'high_load'
            pv_type = 'low_solar'
            t2 += 1
            k = t2
        elif load_batch_mean < load_mean and pv_batch_mean >= pv_mean :
            load_type = 'low_load'
            pv_type = 'high_solar'
            t3 += 1
            k = t3
        else :
            load_type = 'low_load'
            pv_type = 'low_solar'
            t4 += 1
            k = t4
        
        df_merge =  load_batch.merge(pv_batch,on = 'datetime',how = 'inner')
        df_merge.to_csv('batch_dataset/'+pv_type+'_'+load_type+'_' +str(k)+'.csv',index=False)
        dataset_detail.append([pv_type+' '+load_type+'_' +str(k),pv_type,load_type,pv_batch_mean,load_batch_mean,today])
dataset_detail = pd.DataFrame(dataset_detail,columns=['name','pv_type','load_type','pv_batch_mean','load_batch_mean','start_date'])
dataset_detail.to_csv('batch_dataset/' + 'dataset_detail.csv',index=False)
        
```
